```python
#Python program to display daily sales of two salesman with following data
#sales1 = [1, 3, 8, 9, 11]
#sales2 = [2, 5, 9, 11, 13]
#Days = [,3,4,5]
#line plo1,2t

import matplotlib.pyplot as plt
sales1 = [1, 3, 8, 9, 11]
sales2 = [2, 5, 9, 11, 13]
Days = [1,2,3,4,5]
plt.plot(Days, sales1, ".-")
plt.plot(Days, sales2, ":")
plt.title("Sales comparison of two employees")
plt.xlabel("Days")
plt.ylabel("Sales")
plt.legend(['Sales1','Sales2'],loc=4)
plt.show()
```

```python
#Python program to display bar chart with following data
#prog_lanugages = ('Python', 'C++', 'Java', 'Perl', 'C', 'Lisp')
#performance = [10,7,6,4,2,1]
#Bar Plot
import matplotlib.pyplot as plt
import numpy as np
prog_lanugages = ('Python', 'C++', 'Java', 'Perl', 'C', 'Lisp')
a=np.arange(len(prog_lanugages))
performance = [10,7,6,4,2,1]
plt.bar(a,performance,align='center',color='green')
plt.xticks(a,prog_lanugages)
plt.ylabel('Usage')
plt.title("prog_lanugages usage")
plt.show()
```

```python
#Python program to display pie chart with following data
#runs = [77,22,42,103]
#players = ['Rohit sharma','MS dhoni','Jasmit bumrah','Virat kohli']

import matplotlib.pyplot as plt
runs = [77,22,42,103]
players = ['Rohit sharma','MS dhoni','Jasmit bumrah','Virat kohli']
cols = ['c','m','r','g']
plt.pie(runs, labels=players, colors=cols, startangle=90, shadow= True, explode=(0.1,0.1,0.1,0.1), autopct='%.1f%%')
plt.title('Runs Scored Chart')
plt.show()
```
